# importing of required libraries

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("car data.csv")
```

```python
print(df["Fuel_Type"].unique())
print(df["Seller_Type"].unique())
print(df["Owner"].unique())
print(df["Car_Name"].unique()) 
```

### purifing the data set for training 

```python
final_dataset =df[['Year','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner']]
```

```python
final_dataset
```

### finding the current age of car

```python
curr_time = datetime.datetime.now()
#df['age']= df['Year'].apply(lambda x : curr_time.year - x)
final_dataset['age']=final_dataset['Year'].apply(lambda x: curr_time.year-x)
```

```python
final_dataset
```

### droping of extra column

```python
final_dataset=final_dataset.drop(['Year'],axis=1)
```

```python
final_dataset
```

```python
final_dataset=pd.get_dummies(final_dataset,drop_first=True)
```

```python
final_dataset
```

### sepreation of data in input and output feature

```python
X=final_dataset
```

```python
y=df['Selling_Price']

y
```

```python
X
```

### importing of libraries for model building 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

### seperation of data for training and testing phase 

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.2,random_state=2)
```

### scaling of data for good performance

```python
standardScaler = StandardScaler()
standardScaler.fit(X_train)
X_train = standardScaler.transform(X_train)
X_test = standardScaler.transform(X_test)
```

### prediction through linear regression

```python
linearRegression = LinearRegression()
linearRegression.fit(X_train, y_train)
y_pred = linearRegression.predict(X_test)
r2_score(y_test, y_pred)
```

### prediction through random forest regressor

```python
rf = RandomForestRegressor(n_estimators = 1000)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
r2_score(y_test, y_pred)
```

```python
plt.scatter(y_test,y_pred)
plt.show()
```
